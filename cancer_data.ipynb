```python
from sklearn.preprocessing import MinMaxScaler
scaler=MinMaxScaler()
scaler.fit(x_train)
x_train_minmax_scaled=scaler.transform(x_train)
x_test_minmax_scaled=scaler.transform(x_test)
```

```python
knn.fit(x_train_minmax_scaled,y_train)
print("Test set accuracy: {:.2f}".format(knn.score(x_test_minmax_scaled,y_test)))
print("Train set accuracy:{:.2f}".format(knn.score(x_train_minmax_scaled,y_train)))
```

### Testing with different values of k

```python
train_accuracy=[]
test_accuracy=[]

neighbors=range(1,11)

for n in neighbors:
    knn=KNeighborsClassifier(n_neighbors=n)
    knn.fit(x_train_minmax_scaled,y_train)
    train_accuracy.append(knn.score(x_test_minmax_scaled,y_test))
    test_accuracy.append(knn.score(x_train_minmax_scaled,y_train))
    
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
```

### Ploting results of different values of k

```python
plt.plot(neighbors,train_accuracy,label="train_accuracy")
plt.plot(neighbors,test_accuracy,label="test_accuracy")
plt.ylabel("Accuarcy")
plt.xlabel("neighbors")
plt.legend()
```

```python
knn=KNeighborsClassifier(n_neighbors=5)
knn.fit(x_train_minmax_scaled,y_train)
print(knn.score(x_test_minmax_scaled,y_test))
print(knn.score(x_train_minmax_scaled,y_train))
```
